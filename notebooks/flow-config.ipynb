```python
import firedrake as fd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import ufl

import hydrogym as gym

mpl.rc("text", usetex=False)
mpl.rc("font", family="serif")
mpl.rc("xtick", labelsize=14)
mpl.rc("ytick", labelsize=14)
mpl.rc("axes", labelsize=20)
mpl.rc("axes", titlesize=20)
mpl.rc("figure", figsize=(6, 4))
%config InlineBackend.figure_format = 'retina'
```

```python
flow = gym.firedrake.Cylinder(restart='../examples/cylinder/output/checkpoint_medium_p1p1.h5')
fig, ax = plt.subplots(1, 1, figsize=(7.5, 3))
levels = np.linspace(-3, 3, 10)
fd.pyplot.tricontourf(
    flow.vorticity(),
    levels=levels,
    axes=ax,
    extend="both",
    cmap=sns.color_palette("icefire", as_cmap=True),
)
ax.set_xlim([-2, 9])
ax.set_ylim([-2, 2])
ax.set_facecolor("grey")
plt.show()
```

```python
flow = gym.firedrake.envs.Pinball(restart="../examples/pinball/output/checkpoint_fine.h5")

fig, ax = plt.subplots(1, 1, figsize=(7.5, 4.5))
levels = np.linspace(-3, 3, 20)
fd.tricontourf(
    flow.vorticity(),
    levels=levels,
    axes=ax,
    extend="both",
    cmap=sns.color_palette("icefire", as_cmap=True),
)
ax.set_xlim([-2, 12])
ax.set_ylim([-4, 4])
ax.set_facecolor("grey")
plt.show()
```

```python
flow = gym.firedrake.envs.Step(restart="../examples/step/output/checkpoint.h5")

fig, ax = plt.subplots(1, 1, figsize=(12, 2))
levels = np.linspace(-5, 5, 20)
fd.tricontourf(
    flow.vorticity(),
    levels=levels,
    axes=ax,
    extend="both",
    cmap=sns.color_palette("icefire", as_cmap=True),
)
ax.set_xlim([-2, 10])
ax.set_ylim([-0.5, 0.5])
ax.set_facecolor("grey")
plt.show()
```

```python
flow = gym.firedrake.envs.Step(restart="../examples/step/output/checkpoint.h5")
u, p = flow.q.subfunctions[0],flow.q.subfunctions[1]
U = fd.interpolate(ufl.dot(u, u), p.function_space())
fig, ax = plt.subplots(1, 1, figsize=(12, 2))
levels = np.linspace(-1, 1, 20)
fd.tricontourf(
    U,
    levels=levels,
    axes=ax,
    extend="both",
    cmap=sns.color_palette("icefire", as_cmap=True),
)
ax.set_xlim([-2, 10])
ax.set_ylim([-0.5, 0.5])
ax.set_facecolor("grey")
plt.show()
```
